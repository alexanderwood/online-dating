# Virtual Speed Dating: or, Love in the Time of Coronavirus

## Motivation

From online dating to the abundance of smartphone-based dating apps, the way romantic connections are formed underwent major overhauls in the past decades. The answer to the question "how did you first meet" is increasingly through some dating app. In fact, [30% of adults in the United States](https://www.pewresearch.org/internet/2020/02/06/the-virtues-and-downsides-of-online-dating/) have used a dating app - that figure jumps to 48% looking just at ages 19-29, and jumps to 55% looking at lesbian, gay, and bisexual (LGB) adults. Rather than meeting potential partners through existing social ties, people [can connect with complete strangers](https://web.stanford.edu/~mrosenfe/Rosenfeld_How_Couples_Meet_Working_Paper.pdf) instead. 

Enter coronavirus. In the current moment [the coronavirus pandemic has changed the world](https://www.pewsocialtrends.org/dataset/covid-19-late-march-2020/) and, necessarily, the way we date. Since March 12 of this year, [Bumble reports](https://www.motherjones.com/coronavirus-updates/2020/03/dating-apps-bumble-tinder-increase-coronavirus-data/) a 21% increase in messages sent in the United States, surging as high as 26% during lockdown in San Francisco. The apps are emphasizing features that allow folks to connect in socially distanced ways, such as Hinge's [date from home feature](https://www.inputmag.com/culture/hinges-new-feature-will-help-you-avoid-awkward-online-dating-moments), Tinder's [Passport feature](https://www.inputmag.com/tech/tinder-doesn't-want-you-to-be-entirely-alone-during-self-isolationtinder-doesnt-want-you-to-be-lonely-during-quarantine), and Bumble's video chat option.

How can we revolutionize dating for a post-coronavirus society? Plenty of research studies online interactions between users to try and optimize successful romantic connections. However, in the post-coronavirus world, different forms of communicating have taken hold. Folks now commonly partake in virtual happy hours, virtual watch parties, and even video dates.

I suggest we look back to the old school and ask what lessons we can learn from the speed dating event. When is a speed dating even successful, and what kind of personalities perform best at these events? How can these lessons be applied to swipe-based dating? Would a new service, such as virtual speed-dating, be a potential method for deeper emotional connections between users and successful matches?

## Methodology

Apps such as Tinder, Bumble, Hinge, OKCupid, and more all have the same general goal: to facilitate successful romantic connections. Success is a difficult metric to define in such a nebulous scenario, and some artificial metric must be imposed. For example, OKCupid [has performed data analyses](https://theblog.okcupid.com/) with success defined as a certain number of messages being exchanged back and forth between users.

A common method of "matching" with a potential date on these apps has users swipe left to reject another user or right to accept. People are able to message and interact with mutual matches. For heterosexual couples, apps like Bumble require the woman to first accept a message request from any male suitor. On Hinge you don't swipe right, but rather send a "like" on some aspect of a potential date's profile.

Despite variations in matching methods across different apps, swipe-based dating platforms share the same basic steps for introducing matches:
1. Users quickly evaluate each other based on a short profile.
1. Users make a fast decision on whether or not they like the other user.
1. If both users like each other's profiles they are matched. They may proceed to talk and possibly go on an in-person date.

The Speed Dating event can be described with the same general format, with in-person interaction in place of profiles and swipes. Instead of performing evaluation based on other users' profiles, a person decides whether or not they "like" another person by chatting face-to-face with them for a short time span. At the end of the event, matches can choose to exchange information (swiping left) or not (swiping right). 

This format could be adapted for dating apps, having users matched with other users for brief video "dates" and matched based on their decision after a several-minute conversation. If matched, users could then continue to interact "face-to-face" in-app using a video chat feature.

In order to optimize success of a speed dating matching algorithm we need to answer the big question - what drives success in the speed dating format? Is this different from how success is defined in the current swipe-based model? Also - how is success even defined in this scenario? 

Success could be defined in multiple ways:
1. Two participants like each other resulting in a match
1. Even further down the line, as whether or not matched participants went on a date.

**Would a new virtual speed-dating service be a potential method for deeper emotional connections between users and successful matches?** 

### The Dataset

Colombia University professors Ray Fisman and Sheena Iyengar [ran a series of speed dating experiments](http://www.stat.columbia.edu/~gelman/arm/examples/speed.dating/) on Columbia graduate students and have made this data public. They collected details before, during, and after the event, and followed up with participants to evaluate longer-term success rates. All matches were between men and women (in other words, no same-gender matches). 

Twenty-one different speed dating events were held with a total of 552 different individuals attending. Before each event, demographic information such as the subject's field of study, undegraduate alma mater, SAT score, zip code of childhood home, where the subject is from, median household income, race, and gender were collected. A series of preferential questions were also asked, including:
1. How important is it to you (on a scale of 1-10) that a person you date be of the same racial/ethnic background?
1. How important is it to you (on a scale of 1-10) that a person you date be of the same religious background?

Next, questions regarding the expectations of the participants were asked. Before the event, participants were given surveys on the attributes of the types of people they think they would be romantically interested in. They were also given surveys on their attitude about how likely they were to meet someone and general outlook on the event. These questionaires were given before the event, halfway through the event, and after the event.

Full details of survey questions can be seen in the dataset's [data key]().

## Exploratory Analysis

Do participants' attitudes change thorughout the event? How does the initial attitude of a participant effect their ability to go on dates, and how does that compare to their final attitude? Can this initial attitude predict match success? 

To begin to answer these questions, let's start with some exploratory analysis of the data. The first step is to load it up into a Pandas dataframe. 

```python
def load_data():
    df = pd.read_csv("data/dating-data.csv", encoding='latin1')
    # Remove columns with no information.
    drop_cols = [col for col in df.columns if df[col].value_counts().shape[0] <= 1]
    df = df[df.columns.difference(drop_cols)]
    return df

df = load_data()
df.head()
```

There are multiple potential target variables in this data set. A good short-term target is whether or not two participants matched during the event, meaning they chose each other and were eligible to receive each others' contact information. However, I am interested in something deeper: Not just whether participants matched, but whether they *contacted each other*.

So I would like to look at the long-term outcomes. Long-term target variables in this data set, or the outcomes, were collected 3-4 weeks after subjects had been sent their matches. These variables are:
* `you_call`: How many matches has the subject contacted to set up a date?
* `them_cal`: How many have contacted the subject?
* `date_3`: Have you been on a date with any matches? (Yes = 1, no = 2)
* `numdat_3`: How many matches have you been on a date with so far?

We have 195 total columns in the table. Looking at the data key, we see that data includes information such as:
 * `iid`: Unique identifier for each separate subject.
 * `id`: Subject number within a speed dating event, or "wave".
 * `idg`: Subject number within their gender group. 

First let's look at some demographic information. 

```python
ax = df.gender.value_counts().plot.bar()
ax.set_xticklabels(["Women", "Men"], rotation=0)
plt.title("Count of particpants by gender")
```

The number of men and women participating is similar, as expected. 

### Questionaire Answers

Next, let's break down subjects' answers to some of the initial questions they were asked. We can compare responses across gender.

```python
def agg_plot(question, attribute, groupby, answer_key, legend_vals):
    pltdat = df.groupby([groupby]).agg(dat_count=(attribute,'value_counts'))
    pltdat = pltdat.reset_index()

    ax = sns.barplot(x=attribute, y='dat_count', hue=groupby, data=pltdat)
    plt.title(question)
    ax.set_xlabel(attribute)
    ax.set_ylabel('Number Of Occurences')
    ax.set_xticklabels(answer_key, rotation=90)
    
    for a,b in zip(ax.axes.get_legend().texts, legend_vals): a.set_text(b)
        
        
```

```python
question="What is your primary goal in participating in this event?"
answers=['Seemed fun', 'Meet new people', 'Get a date', 'Want LTR', 'To say I did it','Other']
new_legend=['Men','Women']
agg_plot(question, 'goal', "gender", answers, new_legend)
```

The vast majority of participants of both genders participated because it seemed like it would be fun or because they wanted to meet new people. Relatively few participated with the express purpose of finding a date, and even fewer with the goal of finding a long-term relationship.

```python
question = "In general, how frequently do you go on dates?"
answers = ['Several times a week', 'Twice a week', 'Once a week', 'Twice a mont', 'Once a month', 'Several times a year', 'Almost never']
agg_plot(question, 'date', 'gender', answers, new_legend)
```

There's a few superstar daters who go on dates multiple times a week. Most participants claim to go on dates twice a month or less.

Before the event, participants were also asked about what they were looking for in their matches, as well as what they thought their matches would be looking for in them.


```python
question = "What is your intended career?"
answers = ['Lawyer', 'Academic', 'Psychologist', 'Doctor', 'Engineer', 'Entertainment', 'Finance', 'Real Estate', 'International Affairs', 'Undecided', 'Social Work', 'Speech pathology', 'Politics', 'Athletics', 'Other', 'Journalism', 'Architecture']
agg_plot(question, 'career_c', 'gender', answers, new_legend)
```

Unsurprisingly, given the study population, there are a large amount of participants going into academic work. Interestingly, there is a surplus of women participating who are interested in working in the banking/consulting/finance industry. 

Participants also rated the importance of what they are looking for in the opposite sex. Waves 6-9 of participants ranked values on a scale of 1-10; the rest of the 21 waves were given 100 points total to distribute among the categories. To start with, let's look just at the waves that evaluated given the second method of splitting up 100 points between the categories.

```python
def remove_waves(df):
    return df[df.wave.isin(list(range(5)) + list(range(10,22)))]


df2 = remove_waves(df)


def column_names(a, b):
    a, b = str(a), str(b)
    cols = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']
    if a=='3' or a=='4':
        cols = cols[:-1]
    return [c + a + '_' + b for c in cols]
    

def attributes_plot(df, question, a, b, hue_order):  
    categories = ['Attractive', 'Sincere', 'Intelligent', 'Fun', 
                  'Ambitious', 'Shared Interests']
    cols = ['gender'] + column_names(a, b)
    
    # Get average of scores for each gender
    scores = df[cols]
    scores = pd.melt(scores, id_vars=['gender'], value_vars=column_names(a,b))
    ax = sns.catplot(data=scores, x='variable', y='value', hue='gender', hue_order=hue_order,
                kind='violin',split=True, scale='count', height=5, aspect=2, inner='quartile')
    plt.title(question)
    plt.xlabel('Categories')
    plt.ylabel('Importance Level')
    ax.set_xticklabels(categories, rotation=80, ha='center')
    
    legend = ax._legend
    legend_vals = ['Men', 'Women']
    for a,b in zip(legend.texts, legend_vals): a.set_text(b)
        

question = "What are you looking for in the opposite sex?"
attributes_plot(df2,question,1,1, [1, 0])
```

```python
question = "What does the opposite gender look for in a date?"
attributes_plot(df2,question,2,1, [0, 1])
```

Women tend to overestimate how much men care about looks and underestimate how much they care about sincerity and intelligence. Men also overestimate how much women care about look and fun.

Interestingly, participants rank themselves on the first five attributes.

```python
question = 'How do you think you measure up? (After Speed Dating)'
attributes_plot(df2, question, 3, 3, [1, 0])
plt.ylabel("Self-Assessment (1=awful and 10=great)")
```

### Outcomes

There are a number of outcomes that can be assessed with this data. For instance,
1. Number of likes given
1. Number of likes received
1. Number of mutual likes (matches)
1. Initiated contact with match(es) after the event
1. Been on date since the event

These outcomes could be used to select participants with specific behaviors. For instance, a dating app may wish to encourage people who give likes to participate in more events; on the other hand, they may wish to encourage participation more from the type of person who gets likes.

Let's start with number of likes given and see how that is connected to self-assessment before, during, and after the event. Let's remove wave 12, where participants were only allowed to say yes to 50% of participants.

```python
def data_subset(df):
    df = df[df.wave != 12]
    targs = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']
    targs = '|'.join(targs)
    targs = '(' + targs + ')(' + '|'.join([str(x) for x in range(1,10)]) + ')_.*'
    attributes = ['iid', 'gender'] + df.filter(regex=targs, axis=1).columns.to_list()
    print("We will consider {} different predictor variables.".format(len(attributes) - 2))  # don't count gender, id #

    # Set up the target data. Normalize number of likes given by number of potential matches.
    decisions = df[['iid', 'dec']].groupby('iid').sum()
    decisions = decisions.dec/df.iid.value_counts().sort_index()
    
    df2 = df[attributes].drop_duplicates()
    df2 = df2.join(decisions.to_frame(name='like_perc'), on='iid')
    return df2
    
df2 = data_subset(df)
print("Percentage of missing data per variable:")
df2.isna().mean()
```

Out of the predictor variables selected, a fair number of them have no responses. Let's keep only the columns with at least a 75% response rate. We'll remove any remaining empty variable rows for now.

```python
df2 = df2[df2.columns[df2.isnull().mean() < 0.05]].dropna()
print('Pearson correlation between each remaining variable and outcome:')
df2.drop("like_perc", axis=1).apply(lambda x: x.corr(df2.like_perc))
df2.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor


def train_model(X, y, mdl, mdl_name):    
    def visualize_model(X_test, y_test, mdl, mdl_name):
        pred = mdl.predict(X_test)
        fig = plt.figure()
        plt.plot(pred[:40], label=mdl_name)
        plt.plot(y_test[:40], label='Truth')
        plt.legend()
        plt.xlabel('Participant ID')
        plt.ylabel('% of dates given likes')
        plt.title("Predicted Values versus Ground Truth")
      
    
    nfolds = 5
    kf = KFold(n_splits=nfolds)
    score = 0
    for train_idx, test_idx in kf.split(X):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        mdl.fit(X_train, y_train)
        score += mdl.score(X_test, y_test)
    # An example visualization.
    visualize_model(X_test, y_test, mdl, mdl_name)
    score /= nfolds
    print("Score under {}-fold cross validation for {}: {}".format(nfolds, mdl_name, score))
    
    
```

```python
# Split the target and input data.
X = df2.drop(labels=['like_perc', 'gender', 'iid'], axis=1).to_numpy()
y = df2[['like_perc']].to_numpy()

train_model(X, y, LinearRegression(normalize=True), "Linear Regression")
```

Linear regression does not work well for this task. This isn't surprising, given the meager univariate correlation we saw above. 

Let's try Random Forest. I like Random Forest, an ensemble model, because it is better able to identify multivariate patterns in the data and does a respectable job balancing bias and variance.

First thing we notice is that men were far less likely to contact dates.


```python
from sklearn.ensemble import RandomForestRegressor

train_model(X, y.ravel(), RandomForestRegressor(random_state=1, max_depth=4, n_estimators=100, max_features='sqrt'), "Random Forest Regression")
```

This doesn't work as terribly as linear regression - which isn't saying much, because it still doesn't work. It is looking like the number of matches selected is not tied to the attributes that participants say are important to them in potential dates or their own self-assessment of how they measure up in those categories.
